```python
from pyspark.sql import SparkSession
```

```python
from pyspark.ml.regression import LinearRegression
```

```python
spark = SparkSession.builder.appName('Linear Regression').getOrCreate()
```

```python
data = spark.read.format('libsvm').load('sample_linear_regression_data.txt')
```

```python
data.show()
```

```python
linear_reg = LinearRegression(featuresCol='features',labelCol="label", predictionCol="prediction")
```

```python
#split the data into training data and testing data randomly using 7-% and 30% respectively
training_data, testing_data = data.randomSplit([0.7,0.3])
```

```python
training_data.show() #show training data
```

```python
testing_data.show() #show testing data
```

```python
training_data.describe().show() #show attributes of training data
```

```python
testing_data.describe().show() #show attributes of testing data
```

```python
linear_model = linear_reg.fit(training_data) #fit the training data into model
```

```python
result = linear_model.evaluate(testing_data) #evaluate model with test data
```

```python
#Next get values from result
```

```python
result.degreesOfFreedom
```

```python
ulabelled_data = testing_data.select('features') #extract features from test_data
```

```python
ulabelled_data.show()
```

```python
prediction = linear_model.transform(ulabelled_data)
```
